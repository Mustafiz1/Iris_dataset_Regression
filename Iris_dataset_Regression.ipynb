```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 
```

```python
columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class_labels']
```

```python
df = pd.read_csv('iris dataset.csv', names = columns)
```

```python
df.describe()
```

## Data Visualization

```python
sns.pairplot(df, hue = 'class_labels')
```

## Model building

```python
data = df.values
data
```

```python
X = data[:,0:4]
Y = data[:, 4]
```

```python
X[:10]
```

```python
Y[:10]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
print(X_train[:10])
```

```python
X_train.shape
```

```python
X_test.shape
```

## Using Support Vector Machine algorithm

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train, Y_train)
```

```python
pred_svc = model_svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc_svc = accuracy_score(Y_test, pred_svc)
acc_svc
```

```python
for i in range(len(pred_svc)):
    print(Y_test[i], pred_svc[i])
```

## Using Logistic Regression algorithm

```python
from sklearn.linear_model import LogisticRegression

model_LR = LogisticRegression()
model_LR.fit(X_train, Y_train)
```

```python
pred_lr = model_LR.predict(X_test)
```

```python
acc_lr = accuracy_score(Y_test, pred_lr)
acc_lr
```

## Using Decision Tree Classifier algorithm

```python
from sklearn.tree import DecisionTreeClassifier

model_dtc = DecisionTreeClassifier()
model_dtc.fit(X_train, Y_train)
```

```python
pred_dtc = model_dtc.predict(X_test)
```

```python
acc_dct = accuracy_score(Y_test, pred_dtc)
acc_dct
```

## A detailed classification report 

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test, pred_lr))
```
